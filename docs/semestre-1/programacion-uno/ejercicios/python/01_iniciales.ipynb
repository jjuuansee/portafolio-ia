```python
print("NOTA DEL EXAMEN")
print("===============")
respuestas_correctas = int(input("Ingrese la cantidad de respuestas correctas: "))
if respuestas_correctas < 0:
    print("La cantidad de respuestas correctas no puede ser negativa")
else:
    respuestas_incorrectas = int(input("Ingrese la cantidad de respuestas incorrectas: "))
    if respuestas_incorrectas < 0:
        print("La cantidad de respuestas incorrectas no puede ser negativa")
    else:
        respuestas_en_blanco = int(input("Ingrese la cantidad de respuestas en blanco: "))
        if respuestas_en_blanco < 0:
            print("La cantidad de respuestas en blanco no puede ser negativa")
        else:
            puntaje_total = respuestas_correctas + respuestas_incorrectas + respuestas_en_blanco
            print("El puntaje obtenido en el examen es de:", 
                  str(4 * respuestas_correctas - respuestas_incorrectas)+
                  "/"+str(puntaje_total*4))
```

```python
print("POLINOMIO DE SEGUNDO GRADO")
print("==========================")
a = float(input("Ingrese un valor entero de a: "))
if a == 0:
    print("No es una ecuacion de segundo grado")
else:
    b = float(input("Ingrese un valor entero de b: "))
    c = float(input("Ingrese un valor entero de c: "))
    discriminante = b**2 - 4*a*c
    raiz_1 = (-b + discriminante**(1/2)) / (2*a)
    raiz_2 = (-b - discriminante**(1/2)) / (2*a)
    print("Ra√≠z 1:", raiz_1)
    print("Ra√≠z 2:", raiz_2)
```

```python
print("CALCULADORA DE SEGUNDOS EN DETERMINADO TIEMPO")
print("=============================================")
segundos = int(input("Ingrese el de segundos: "))
minutos = int(input("Ingrese el total de minutos: "))
horas = int(input("Ingrese el total de horas: "))
tiempo_total = horas*(60**2)+minutos*60+segundos
print("El total de segundos ingresados es de", tiempo_total)
```
