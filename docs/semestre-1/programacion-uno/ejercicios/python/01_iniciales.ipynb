### Ejercicio: Identificaci√≥n

### Ejercicio: Horario (hh:mm:ss)

**Objetivo:** solicitar horas, minutos y segundos y mostrar el horario en formato `hh:mm:ss`.

**Entradas:** valores como cadenas o n√∫meros; en este ejemplo se concatenan cadenas.

**Consideraciones:** no hay validaci√≥n de rango (0‚Äì23 para horas, 0‚Äì59 para minutos/segundos); como ejercicio adicional puedes agregar comprobaciones y formateo con ceros a la izquierda.

```python
print("HORARIO ESCRITO EN FORMATO hh:mm:ss")
print("===================================")
horas = input("Ingrese la horas: ")
minutos = input("Ingrese los minutos: ")
segundos = input("Ingrese los segundos: ")
print("El horario es: "+horas+":"+minutos+":"+segundos)
```

### Ejercicio: C√°lculo de nota del examen
**Objetivo:** calcular el puntaje final teniendo en cuenta respuestas correctas (+4), incorrectas (-1) y en blanco (0), y mostrar el resultado sobre el puntaje m√°ximo.

**Entradas:** enteros para respuestas correctas, incorrectas y en blanco. Se validan valores negativos.

**Consideraciones:** el puntaje total se calcula como `4*correctas - incorrectas` y la nota se muestra como `obtenido/total_posible`.

```python
print("NOTA DEL EXAMEN")
print("===============")
respuestas_correctas = int(input("Ingrese la cantidad de respuestas correctas: "))
if respuestas_correctas < 0:
    print("La cantidad de respuestas correctas no puede ser negativa")
else:
    respuestas_incorrectas = int(input("Ingrese la cantidad de respuestas incorrectas: "))
    if respuestas_incorrectas < 0:
        print("La cantidad de respuestas incorrectas no puede ser negativa")
    else:
        respuestas_en_blanco = int(input("Ingrese la cantidad de respuestas en blanco: "))
        if respuestas_en_blanco < 0:
            print("La cantidad de respuestas en blanco no puede ser negativa")
        else:
            puntaje_total = respuestas_correctas + respuestas_incorrectas + respuestas_en_blanco
            print("El puntaje obtenido en el examen es de:", 
                  str(4 * respuestas_correctas - respuestas_incorrectas)+
                  "/"+str(puntaje_total*4))
```

### Ejercicio: Polinomio de segundo grado

**Objetivo:** calcular las ra√≠ces de una ecuaci√≥n cuadr√°tica `ax^2 + bx + c = 0` usando la f√≥rmula general.

**Entradas:** coeficientes `a`, `b`, `c` (con `a != 0`).

**Consideraciones:** si `a == 0` no es una ecuaci√≥n de segundo grado; el discriminante `b^2-4ac` determina si las ra√≠ces son reales. Como ejercicio, manejar casos de discriminante negativo o usar n√∫meros complejos.

```python
print("POLINOMIO DE SEGUNDO GRADO")
print("==========================")
a = float(input("Ingrese un valor entero de a: "))
if a == 0:
    print("No es una ecuacion de segundo grado")
else:
    b = float(input("Ingrese un valor entero de b: "))
    c = float(input("Ingrese un valor entero de c: "))
    discriminante = b**2 - 4*a*c
    raiz_1 = (-b + discriminante**(1/2)) / (2*a)
    raiz_2 = (-b - discriminante**(1/2)) / (2*a)
    print("Ra√≠z 1:", raiz_1)
    print("Ra√≠z 2:", raiz_2)
```

### Ejercicio: Calculadora de segundos

**Objetivo:** convertir una cantidad de horas, minutos y segundos a un √∫nico valor en segundos.

**Entradas:** enteros para segundos, minutos y horas.

**Salida:** n√∫mero entero con el total de segundos (`horas*3600 + minutos*60 + segundos`).

**Consideraciones:** se asume que los valores son no negativos; como mejora se puede validar la entrada y permitir valores grandes.

```python
print("CALCULADORA DE SEGUNDOS EN DETERMINADO TIEMPO")
print("=============================================")
segundos = int(input("Ingrese el de segundos: "))
minutos = int(input("Ingrese el total de minutos: "))
horas = int(input("Ingrese el total de horas: "))
tiempo_total = horas*(60**2)+minutos*60+segundos
print("El total de segundos ingresados es de", tiempo_total)
```
