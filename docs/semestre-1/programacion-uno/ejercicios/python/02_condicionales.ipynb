### Ejercicio: Año bisiesto

**Objetivo:** pedir un año, y determinar si el año ingresado es bisiesto.

**Entradas:** número entero (año).

**Salida:** impresión por pantalla.

**Consideraciones:** se puede probar con años muy futuros como 30990.

```python
print("AÑO BISIESTO")
print("^^^^^^^^^^^^")
año = int(input("Ingrese el año:"))
if año < 0:
    print("Ingrese un año válido")
elif año % 4 == 0:
    print("El año es bisiesto")
elif año % 4 == 0 and año % 100 == 0:
    print("El año es bisiesto")
else:
    print("El año no es bisiesto")
```

### Ejercicio: Horas anteriores y posteriores

**Objetivo:** solicitar dos instantes de tiempo (hora, minuto, segundo) y determinar si la primera hora es anterior, igual o posterior a la segunda.

**Entradas:** enteros para horas (0–23), minutos (0–59) y segundos (0–59) para cada instante.

**Salida:** mensaje indicando si la primera hora es anterior, igual o posterior.

**Consideraciones:** validar rangos numéricos; tratar correctamente igualdad y comparar horas, luego minutos y segundos en ese orden.

```python
print("HORAS ANTERIORES Y POSTERIORES")
print("==============================")
hora1 = int(input("Ingrese la primer hora: "))
hora2 = int(input("Ingrese la segunda hora: "))
if hora1 < 0 or hora1 > 23 or hora2 < 0 or hora2 > 23:
    print("Las horas ingresadas son invalidas")
else:
    minuto1 = int(input("Ingrese los minutos de la primer hora: "))
    minuto2 = int(input("Ingrese los minutos de la segunda hora: "))
    if minuto1 < 0 or minuto1 > 59 or minuto2 < 0 or minuto2 > 59:
        print("Los minutos ingresados son invalidos")
    else:
        segundo1 = int(input("Ingrese los segundos de la primer hora: "))
        segundo2 = int(input("Ingrese los segundos de la segunda hora: "))
        if segundo1 < 0 or segundo1 > 59 or segundo1 < 0 or segundo1 > 59:
            print("Los segundos ingresados son invalidos")
        elif hora1 < hora2:
            print("La primer hora ingresada es anterior a la segunda hora")
        elif hora1 == hora2 and minuto1 < minuto2:
            print("La primer hora ingresada es anterior a la segunda hora")
        elif hora1 == hora2 and minuto1 == minuto2 and segundo1 < segundo2:
            print("La primer hora ingresada es anterior a la segunda hora")
        elif hora1 == hora2 and minuto1 == minuto2 and segundo1 == segundo2:
            print("La primer hora ingresada es igual a la segunda hora")
        else:
            print("La primer hora ingresada es posterior a la segunda hora ingresada")
```

### Ejercicio: Número dentro de un rango

**Objetivo:** comprobar si un número `n` está entre dos límites `linf` y `lsup` 
(inclusive).

**Entradas:** `n`, `linf`, `lsup` (enteros).

**Salida:** `Verdadero` si `linf <= n <= lsup`, `Falso` en caso contrario.

**Consideraciones:** comprobar que `linf <= lsup` o manejar intercambios.

```python
print("ALGORIMO PARA SABER SI UN NUMERO ESTA DENTRO DE OTROS DOS NUMEROS")
print("=================================================================")
n = int(input("Ingrese el numero que desea saber si esta entro otros dos numeros: "))
linf = int(input("Ingrese el numero inferior:"))
lsup = int(input("Ingrese el numero superior:"))
if linf <= n <= lsup:
    print("Verdadero")
else:
    print("Falso")
```

**Objetivo:** verificar si una fecha (día, mes, año) es válida teniendo en cuenta meses con 30/31 días y años bisiestos.

**Entradas:** `dia`, `mes`, `año` (enteros).

**Salida:** indicar si la fecha es válida o no.

**Consideraciones:** manejar bisiestos correctamente y validar rangos (mes 1–12, día acorde al mes).

```python
print("INDICAR SI UNA FECHA ES VALIDA")
print("==============================")
dia = int(input("Ingrese el dia correspondiente para la fecha: "))
mes = int(input("Ingrese el mes correspondiente para la fecha: "))
año = int(input("Ingrese el año correspondiente para la fecha: "))
es_valido = True
if año < 0:
    print("Ingrese un año válido")
elif año % 4 == 0:
    bisiesto = True
elif año % 4 == 0 and año % 100 == 0:
    bisiesto = True
else:
    bisiesto = False
if dia > 31:
    es_valido = False
elif mes == 4 or mes == 6 or mes == 9 or mes == 11 and dia > 30:
    es_valido = False
elif bisiesto == True and mes == 2 and dia > 29 or bisiesto == False and mes == 2 and dia > 28:
    es_valido = False
else:
    es_valido = True
if es_valido:
    print("La fecha", str(dia)+"/"+str(mes)+"/"+str(año), "es valida")
else:
    print("La fecha", str(dia)+"/"+str(mes)+"/"+str(año), "no es valida")
```

### Ejercicio: Calculadora de IMC

**Objetivo:** calcular el índice de masa corporal (IMC) y clasificar el resultado.

**Entradas:** `peso` (kg) y `altura` (m).

**Salida:** clasificación: Obesidad, Sobrepeso, Peso adecuado, Bajo peso.

**Consideraciones:** validar altura y peso; usar la fórmula clásica `peso / altura**2` y evitar exponente 2.2 sin justificar; redondear de forma adecuada.

```python
print("CALCULADORA DE IMC")
print("==================")
peso = float(input("Ingrese su peso en kg: "))
altura = float(input("Ingrese su altura en metros: "))
if altura > 9.9:
    print("Ingrese una altura valida")
else:
    imc = round(peso / altura ** 2.2)
    if imc >= 30:
        print("Obesidad")
    elif 25 <= imc <= 29.9:
        print("Sobrepeso")
    elif 18.5 <= imc <= 24.9:
        print("Peso adecuado")
    elif imc < 18.5:
        print("Bajo peso")
```

### Ejercicio: Pago por peso del paquete

**Entradas:** `peso` en kg (float).

### Ejercicio: Calculadora de sueldo de una empresa

**Objetivo:** calcular el sueldo semanal aplicando horas extra al doble del pago por hora si las horas superan 40.

**Entradas:** `horas_trabajadas` (entero) y `paga_por_hora` (entero/float).

**Salida:** sueldo total de la semana.

**Consideraciones:** validar horas y pago por hora; asegurar que las horas extra se calculen correctamente.

```python
print("CALCULADORA DE SUELDO DE UNA EMPRESA")
print("====================================")
horas_trabajadas = int(input("Ingrese la horas trabajadas en la semana:"))
paga_por_hora = int(input("Ingrese lo que te pagan por hora:"))
if horas_trabajadas > 40:
    sueldo_total = ((horas_trabajadas-40)*(2*paga_por_hora))+(40*paga_por_hora)
else:
    sueldo_total = horas_trabajadas*paga_por_hora
print("Tu sueldo total de esta semana es de: "+str(sueldo_total)+"$")
```
