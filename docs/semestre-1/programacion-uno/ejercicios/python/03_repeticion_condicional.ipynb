```python
print("SUMATORIA HASTA INGRESAR 0")
print("==========================")
sumatoria = 0
seguir_leyendo = True
while seguir_leyendo:
    numero = int(input("Ingrese un numero:"))
    if numero == 0:
        seguir_leyendo = False
    else:
        sumatoria = numero + sumatoria
print("Su sumatoria es ", sumatoria)
```

```python
print("LEER NUMEROS HASTA QUE SE INGRESE 0 Y DECIR CUAL FUE EL MAYOR INGRESADO")
print("=======================================================================")
mayor = -999999999999
seguir_leyendo = True
while seguir_leyendo:
    numero = int(input("Ingrese un numero: "))
    if numero < 0:
        print("El numero ingresado es invalido")
        seguir_leyendo = False
    elif numero > mayor:
        mayor = numero
    elif numero == 0:
        seguir_leyendo = False
        print("El mayor es ", mayor)
```

```python
print("DIBUJAR UNA FIGURA")
print("==================")
altura = int(input("Ingrese la altura de la figura dibujar:"))
ancho = int(input("Ingrese el ancho de la figura que quiere dibujar:"))
if altura < 1:
    print("La altura ingresada no es correcta")
elif altura == 1:
    print("x"*ancho)
else:
    contador = altura
    print("x"*ancho)
    while contador > 2:
        print("x"+" "*(ancho-2)+"x")
        contador -= 1
    print("x"*ancho)
```

```python
print("CANTIDAD DE NUMEROS QUE SALIERON AL TIRAR UN DADO")
print("=================================================")
import random
def tirada_dados():
    return random.randint(1,6)
num_tiradas = 300000
contador_de_1 = 0
contador_de_2 = 0
contador_de_3 = 0
contador_de_4 = 0
contador_de_5 = 0
contador_de_6 = 0
for i in range(num_tiradas):
    dado = tirada_dados()
    if dado == 1:
        contador_de_1 += 1
    elif dado == 2:
        contador_de_2 += 1
    elif dado == 3:
        contador_de_3 += 1
    elif dado == 4:
        contador_de_4 += 1
    elif dado == 5:
        contador_de_5 += 1
    else:
        contador_de_6 += 1
print("La cantidad de caras 1, salio", contador_de_1)
print("La cantidad de caras 2, salio", contador_de_2)
print("La cantidad de caras 3, salio", contador_de_3)
print("La cantidad de caras 4, salio", contador_de_4)
print("La cantidad de caras 5, salio", contador_de_5)
print("La cantidad de caras 6, salio", contador_de_6)
```

```python
print("PROMEDIO DE VALORES DE LAS TIRADAS DE DOS DADOS")
print("===============================================")
import random
def tirada_dados():
    return random.randint(1,6)
num_tiradas = 20
suma_resultados = 0
for i in range(num_tiradas):
    dado1 = tirada_dados()
    print(dado1)
    dado2 = tirada_dados()
    print(dado2)
    suma_resultados = suma_resultados + dado1 + dado2
promedio = suma_resultados / num_tiradas*2
print("El promedio de tiradas de los dados es:", promedio)
```

```python
print("NUMERO DE LA SUERTE")
print("===================")

def ultimodigito(n):
    return n % 10

def sacarultimodigito(n):
    return n // 10

def suma_digitos(numero):
    suma = 0
    while numero > 0:
        suma += ultimodigito(numero)
        numero = sacarultimodigito(numero)
    return suma


numero_ingresado = int(input("Ingresa un número entero positivo mayor que 0: "))
if numero_ingresado <= 0:
    print("El número debe ser mayor que 0.")
else:
    resultado = suma_digitos(numero_ingresado)
    print(f"El numero {numero_ingresado} es de la suerte: {True if resultado == 21 else False}")

```

```python
print("SUMA DE LOS DIGITOS DE UN NUMERO")
print("================================")

def ultimodigito(n):
    return n % 10

def sacarultimodigito(n):
    return n // 10

def suma_digitos(numero):
    suma = 0
    while numero > 0:
        suma += ultimodigito(numero)
        numero = sacarultimodigito(numero)
    return suma


numero_ingresado = int(input("Ingresa un número entero positivo mayor que 0: "))
if numero_ingresado <= 0:
    print("El número debe ser mayor que 0.")
else:
    resultado = suma_digitos(numero_ingresado)
    print(f"La suma de los dígitos de {numero_ingresado} es: {resultado}")
```

```python
print("SUMATORIA EXTRANA")
print("=================")

k = int(input("Ingrese un numero entero:"))
contador = 0
while contador <= k:
    sumatoria = ((1/16)**k)*((4/8*k+1)-(2/8*k+4)-(1/8*k+5)-(1/8*k+6))
    contador += 1
print(f"La sumatoria total es:{sumatoria}")
```

```python
print("ALGORITMO DE HAILSTONE")
print("======================")
n = int(input("Ingrese un numero entero:"))
while n > 1:
    if n % 2 == 0:
        n = n / 2
    else:
        n = n*3 + 1
    print(int(n))
```
