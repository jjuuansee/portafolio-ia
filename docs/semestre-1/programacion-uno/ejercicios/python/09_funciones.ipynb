### Ejercicio: Mayor de tres n칰meros

**Objetivo:** crear una funci칩n que reciba tres n칰meros enteros y determine cu치l es el mayor de ellos.

**Entradas:** tres n칰meros enteros.

**Salida:** el valor m치ximo encontrado.

```python
print("QUE NUMERO ES MAYOR")
print("===================")


def mayor_de_tres_numeros(n1, n2, n3):
    global es_mayor
    if n1 > n2 and n1 > n3:
        es_mayor = n1
    elif n2 > n1 and n2 > n3:
        es_mayor = n2
    elif n3 > n1 and n3 > n2:
        es_mayor = n3
    elif n1 == n2 == n3:
        print("Los numeros son iguales")
    return es_mayor


n1 = int(input("Ingrese un numero entero: "))
n2 = int(input("Ingrese otro numero entero: "))
n3 = int(input("Ingrese otro numero entero: "))
solucion = mayor_de_tres_numeros(n1, n2, n3)
print("El mayor de los numeros ingresados es:", solucion)
```

### Ejercicio: Dibujar rect치ngulo hueco

**Objetivo:** implementar un procedimiento que imprima en pantalla un rect치ngulo hueco de caracteres 'x' seg칰n las dimensiones dadas.

**Entradas:** `ancho` y `altura` (enteros).

**Salida:** dibujo en consola.

```python
print("DIBUJA UNA FORMA")
print("================")

def dibuja_forma(ancho,altura):
    if altura < 1:
        print("La altura ingresada no es correcta")
    elif altura == 1:
        print("x" * ancho)
    else:
        contador = altura
        print("x" * ancho)
        while contador > 2:
            print("x" + " " * (ancho - 2) + "x")
            contador -= 1
        print("x" * ancho)


altura1 = int(input("Ingrese la altura de la figura dibujar:"))
ancho1 = int(input("Ingrese el ancho de la figura que quiere dibujar:"))
dibuja_forma(ancho1, altura1)
```

### Ejercicio: 칈ndice de Masa Corporal

**Objetivo:** calcular el IMC dado el peso y la altura de una persona (`peso / altura^2`).

**Entradas:** peso (kg) y altura (m).

**Salida:** valor del IMC (redondeado).

```python
print("INDICE DE MASA CORPORAL")
print("=======================")


def masa_corporal(peso,altura):
    imc = round(peso / altura ** 2.2)
    return imc


peso_ej = float(input("Ingrese su peso en kg: "))
altura_ej = float(input("Ingrese su altura en metros: "))
imc = masa_corporal(peso_ej,altura_ej)
print(f"Tu indice de masa corporal es: {imc}")
```

### Ejercicio: F칩rmula cuadr치tica

**Objetivo:** calcular las dos ra칤ces de una ecuaci칩n de segundo grado utilizando la f칩rmula general.

**Entradas:** coeficientes `a`, `b`, `c`.

**Salida:** tupla con las dos ra칤ces.

```python
print("POLINOMIO DE SEGUNDO GRADO")
print("==========================")

import math


def calcular_raices(a, b, c):
    discriminante = b ** 2 - 4 * a * c
    raiz1 = (-b + math.sqrt(discriminante)) / (2 * a)
    raiz2 = (-b - math.sqrt(discriminante)) / (2 * a)
    return raiz1, raiz2


a = int(input("Ingrese el valor de a: "))
b = int(input("Ingrese el valor de b: "))
c = int(input("Ingrese el valor de c: "))
raiz1, raiz2 = calcular_raices(a, b, c)
print(f"Ra칤z 1: {raiz1}")
print(f"Ra칤z 2: {raiz2}")
```

### Ejercicio: Diferencia de tiempos

**Objetivo:** calcular la cantidad total de segundos de dos instantes de tiempo y determinar la diferencia absoluta entre ellos.

**Entradas:** horas, minutos y segundos para dos instantes.

**Salida:** diferencia en segundos.

```python
print("TOTAL DE SEGUNDOS Y DIFERENCIA DE SEGUNDOS")
print("==========================================")


def total_de_segundos(horas,minutos,segundos):
    tiempo_total = horas * (60 ** 2) + minutos * 60 + segundos
    return tiempo_total


horas1 = int(input("Ingrese el numero de horas: "))
minutos1 = int(input("Ingrese el numero de minutos: "))
segundos1 = int(input("Ingrese el numero de segundos: "))
tiempo_total1 = total_de_segundos(horas1,minutos1,segundos1)
horas2 = int(input("Ingrese el numero de la segunda hora: "))
minutos2 = int(input("Ingrese el numero de los segundos minutos: "))
segundos2 = int(input("Ingrese el numero de la tercer hora: "))
tiempo_total2 = total_de_segundos(horas2,minutos2,segundos2)
diferencia = horas2 - horas1
if diferencia > 0:
    print(f"La diferencia de segundos es de: {diferencia}")
else:
    print(f"La diferencia de segundos es de: {diferencia * -1} segundo")
```

### Ejercicio: Secuencia de Hailstone (Collatz)

**Objetivo:** implementar la funci칩n que genera e imprime la secuencia de Collatz para un n칰mero dado hasta llegar a 1.

**Entradas:** n칰mero inicial.

**Salida:** impresi칩n de los t칠rminos de la secuencia.

```python
print("FUNCION DE HAILSTONE")
print("====================")


def hailstone(n):
    while n > 1:
        if n % 2 == 0:
            n = n / 2
        else:
            n = n * 3 + 1
        print(int(n))
    return "FIN"


n = int(input("Ingrese un numero entero: "))
print(hailstone(n))
```
