### Ejercicio: Suma de cr√©ditos universitarios

**Objetivo:** recorrer un diccionario que asocia materias con sus cr√©ditos y calcular el total acumulado.

**Entradas:** diccionario precargado.

**Salida:** listado de materias con sus cr√©ditos y la suma total.

```python
creditos = {
    'Matem√°ticas': 6,
    'F√≠sica': 4,
    'Qu√≠mica': 5,
    'Contabilidad': 8,
    'Programaci√≥n': 6,
    'Redacci√≥n': 6,
    'Trabajo final': 6
}
total_creditos = 0
for materia, credito in creditos.items():
    print(f'{materia} tiene: {credito} creditos')
    total_creditos += credito
print(f"El total de creditos es de: {total_creditos}")
```

### Ejercicio: Consulta de divisas

**Objetivo:** permitir al usuario ingresar el nombre de una divisa y mostrar su s√≠mbolo correspondiente buscando en un diccionario.

**Entradas:** nombre de la divisa (string).

**Salida:** s√≠mbolo de la divisa o mensaje de error si no existe.

```python
diccionario = {
    "Euro": "‚Ç¨",
    "Dolar": "U$D",
    "Yen": "¬•",
    "Peso": "$",
    "Real": "R$"
}
accion = input("Ingrese la divisa que quiere saber su simbolo: ").title()
if accion in diccionario.keys():
    print(diccionario[accion])
else:
    print("Esa divisa no se encuentra en el diccionario")
```
