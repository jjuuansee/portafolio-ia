# Parcial - Juegos Olímpicos

```python
class Deportista:
    def __init__(self, nombre, pais, puntuacion):
        self.nombre = nombre
        self.pais = pais
        self.__puntuacion = puntuacion
        self.deporte = None

    def participar(self):
        raise NotImplementedError("Este método debe ser sobrescrito por cada deporte.")

    def obtener_puntuacion(self):
        print(f'La puntuación es de {self.__puntuacion}')

    def clasificar(self):
        print(f'El atleta {self.nombre} obtuvo {self.__puntuacion} puntos.')

class JuezOlimpico:
    def __init__(self, nombre):
        self.nombre = nombre

    def evaluar_competencia(self, deportistas):
        for atleta in deportistas:
            if atleta._Deportista__puntuacion > 80:  # Se accede al atributo privado
                print(f'El atleta {atleta.nombre} ha clasificado.')
            else:
                print(f'El atleta {atleta.nombre} no ha clasificado.')

class JuegosOlimpicos:
    def __init__(self):
        self.atletismo = []
        self.natacion = []
        self.ciclismo = []

    def agregar_deportista(self, deportista):
        if deportista.deporte == 'Corredor':
            self.atletismo.append(deportista)
        elif deportista.deporte == 'Nadador':
            self.natacion.append(deportista)
        elif deportista.deporte == 'Ciclista':
            self.ciclismo.append(deportista)

class Corredor(Deportista):
    def __init__(self, nombre, pais, puntuacion):
        super().__init__(nombre, pais, puntuacion)
        self.deporte = 'Corredor'

    def participar(self, metros, tiempo):
        self.__puntuacion = metros / tiempo
        print(f'El corredor {self.nombre} corrió {metros} metros en {tiempo} segundos.')

class Nadador(Deportista):
    def __init__(self, nombre, pais, puntuacion):
        super().__init__(nombre, pais, puntuacion)
        self.deporte = 'Nadador'

    def participar(self, metros, tiempo):
        self.__puntuacion = metros / tiempo
        print(f'El nadador {self.nombre} nadó {metros} metros en {tiempo} segundos.')

class Ciclista(Deportista):
    def __init__(self, nombre, pais, puntuacion):
        super().__init__(nombre, pais, puntuacion)
        self.deporte = 'Ciclista'

    def participar(self, kilometros, tiempo):
        self.__puntuacion = kilometros / tiempo
        print(f'El ciclista {self.nombre} recorrió {kilometros} kilómetros en {tiempo} horas.')

# Crear deportistas
corredor1 = Corredor("Usain Bolt", "Jamaica", 0)
nadador1 = Nadador("Michael Phelps", "EE.UU.", 0)
ciclista1 = Ciclista("Chris Froome", "Reino Unido", 0)

# Participación de los deportistas
corredor1.participar(metros=100, tiempo=9.58)
nadador1.participar(metros=200, tiempo=110)
ciclista1.participar(kilometros=50, tiempo=2)

# Crear los Juegos Olímpicos y agregar a los deportistas
juegos = JuegosOlimpicos()
juegos.agregar_deportista(corredor1)
juegos.agregar_deportista(nadador1)
juegos.agregar_deportista(ciclista1)

# Mostrar las listas de participantes
print("Atletismo:", [deportista.nombre for deportista in juegos.atletismo])
print("Natación:", [deportista.nombre for deportista in juegos.natacion])
print("Ciclismo:", [deportista.nombre for deportista in juegos.ciclismo])

# Crear un juez y evaluar la competencia
juez = JuezOlimpico("Juez Internacional")
juez.evaluar_competencia([corredor1, nadador1, ciclista1])

# Mostrar puntuaciones
corredor1.obtener_puntuacion()
nadador1.obtener_puntuacion()
ciclista1.obtener_puntuacion()

# Clasificar los deportistas
corredor1.clasificar()
nadador1.clasificar()
ciclista1.clasificar()
```
