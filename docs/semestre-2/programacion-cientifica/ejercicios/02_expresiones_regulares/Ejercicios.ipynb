### Extracción de datos entre paréntesis

```python
texto = 'Los resultados (2023) fueron mejores que los del año anterior (2022).'
print(re.findall(r'\((.*?)\)', texto))
# \( y \) busca los parentesis
# (.*?) Coincide con todos los caracteres dentro del parentesis y deja de coincidir con
# caracteres cuando se cierra el parentesis. A esto se le llama busqueda codiciosa
```

### Extracción de IP

```python
texto = 'El servidor 192.168.1.1 respondió, pero 10.0.0.5 no esta disponible'
print(re.findall(r'\d+\.\d+\.\d+\.\d+', texto))
```

### Extracción de números

```python
texto = 'En el año 2023, compré 3 libros por 15$ cada uno.'
print(re.findall(r'\d+', texto))
# \d+ encuentra todos los números en el texto
```

### Números de teléfono

```python
def validar_telefono(numero):
    if re.match(r'^\+?(\d{1,3})?\s?\d{9,12}$', numero):
        # ^ indica el inicio de la cadena
        # \+? Puede comenzar con un + o no
        # (\d{1,3})? Puede cotener un grupo de 1 a 3 digitos
        # \s? Puede tener un espacio
        # \d{9,12}$ Hay 9 a 12 digitos despues de todo lo anterior
        print('Número válido')
    else:
        print('Número inválido')

Valido1 = '+598 091255770'
Valido2 = '34 678123456'
Valido3 = '123456789012'
NoValido = '12345'
NoValido2 = '+123456'
NoValido3 = '+12 345'

validar_telefono(Valido1)
validar_telefono(Valido2)
validar_telefono(Valido3)
validar_telefono(NoValido)
validar_telefono(NoValido2)
validar_telefono(NoValido3)
```

### Validación de URL

```python
def validarURL(url):
    if re.match(r'^https?://(www\.)?[\w-]+\.\w+', url):
        print('Url valida')
    else:
        print('Url invalida')
        
ejemploValido = 'https://www.example.com'
ejemploValido2 = 'http://example.com'
ejemploValido3 = 'http://example.com'

ejemploNoValido = 'http://'
ejemploNoValido2 = '://example.com'
ejemploNoValido3 = 'example.'

validarURL(ejemploValido)
validarURL(ejemploValido2)
validarURL(ejemploValido3)
print('\n')

validarURL(ejemploNoValido)
validarURL(ejemploNoValido2)
validarURL(ejemploNoValido3)
```

### Validación de contraseñas

```python
def validador_de_contraseñas(contraseña):
    if re.search(r'[0-9]',contraseña) and re.search(r'[A-Z]',contraseña) and re.search(r'[a-z]',contraseña) and re.search(r'[@#$%^&*()!]',contraseña):
        print('Contraseña valida')
    else:
        print('Contraseñas no valida')

contrasenaValida = 'Ju4n$e'
contrasenaValida1 = 'Segura123!'
contrasenaNoValida = 'Juanse!'
contrasenaNoValida2 = 'Pa$$word'
(validador_de_contraseñas(contrasenaValida))
(validador_de_contraseñas(contrasenaValida1))
(validador_de_contraseñas(contrasenaNoValida))
(validador_de_contraseñas(contrasenaNoValida2))
```

### Validación de fechas

```python
fecha = '29/08/2024'
fecha2 = '15/02/2006'
fecha3 = '32/05/2021'
fecha4 = '12/13/2021'

def validar_fecha(fecha):
    if re.match(r'^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[0-2])/\d{4}$', fecha):
        return 'Fecha valida'
    return 'Fecha invalida'


print(validar_fecha(fecha))
print(validar_fecha(fecha2))
print(validar_fecha(fecha3))
print(validar_fecha(fecha4))
```

### Verificación de correo

```python
import re

correo_bien = 'juanparoli14@gmail.com'
correo_mal = ('hola{Mundo}@pedrito.com')


def verificar_correo(correo):
    if re.match(r'[\w-]+@[\w-]+\.[\w-]{3}', correo):
        return("Correo válido")
    else:
        return("Correo inválido")


print(verificar_correo(correo_bien))
print(verificar_correo(correo_mal))
```
