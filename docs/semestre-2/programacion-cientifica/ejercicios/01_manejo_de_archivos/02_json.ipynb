# Manejo de Archivos JSON

## 1. Crear archivo JSON (Libro)

```python
import json
diccionario = {
    "Libro": [
        {
            "Nombre": "Fahrenheit 451", 
            "Autor": "Ray Bradbury", 
            "Publicacion": "1953", 
            "Tipo": "Novela"
            },
        {
            "Nombre": "La divina comedia", 
            "Autor": "Dante Alighieri", 
            "Publicacion": "Desconocido", 
            "Tipo": "Poema"
            }
    ]
}

with open('data/libro.json', mode='w', encoding='utf-8') as archivo:
    json.dump(diccionario, archivo, indent=4)
```

## 2. Filtrar empleados por salario

```python
import json
mayores_50k = []

with open('data/Empleados.json', mode='r', encoding='utf-8') as archivo:
    datos_empleados = json.load(archivo)
    for empleado in datos_empleados:
        if empleado['salario'] >= 50000:
            mayores_50k.append(empleado)
            
with open('data/Empleados50k.json', mode='w', encoding='utf-8') as archivo:
    json.dump(mayores_50k, archivo, ensure_ascii=False, indent=4)
```

## 3. Convertir CSV a JSON

```python
import csv
import json

datos_csv = []
with open('data/estudiante.csv', mode='r', encoding='utf-8') as archivo:
    lector_csv = csv.DictReader(archivo)
    filas = list(lector_csv)
with open('data/estudiante_conv.json', mode='w', encoding='utf-8') as archivo:
    json.dump(filas, archivo, ensure_ascii=False, indent=4)
```
