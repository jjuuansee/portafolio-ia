# Manejo de Archivos CSV

## 1. Leer archivo CSV (Estudiantes)

```python
import csv
with open('data/estudiante.csv', mode='r', encoding='utf-8') as archivo:
    contenido = list(csv.reader(archivo))
    encabezados = contenido[0]
    for fila in contenido[1:]:
        resultado = f"{encabezados[0]}: {fila[0]} | {encabezados[1]}: {fila[1]} | {encabezados[2]}: {fila[2]}"
        print(resultado)
```

## 2. Calcular promedio de notas

```python
import csv

with open('data/Notas.csv', mode='r', encoding='utf-8') as notas:
    notas_alumnos = csv.reader(notas)
    encabezado = next(notas_alumnos)
    promedio = 0
    cantidad_alumnos = 0
    for nota in notas_alumnos:
        promedio += int(nota[1])
        cantidad_alumnos += 1
    promedio = promedio / cantidad_alumnos
    print(f'El promedio de la clase es: {promedio}')
```

## 3. Filtrar estudiantes mayores de 15

```python
import csv

mayores = []
with open('data/estudiante.csv', mode='r', encoding='utf-8') as archivo:
    contenido = list(csv.reader(archivo, delimiter=','))
    for fila in contenido:
        if fila[1] >= '15':
            mayores.append(fila)

with open('data/mayores_15.csv', mode='w', encoding='utf-8') as archivo:
    writer = csv.writer(archivo, delimiter=',')
    for fila in mayores:
        writer.writerow(fila)
```

## 4. Procesar ventas y exportar a JSON/XML

```python
import csv
import json
import xml.etree.ElementTree as ET

with open('data/Ventas.csv', mode='r', encoding='utf-8') as archivo:
    archivo = csv.reader(archivo)
    encabezado = next(archivo)

    total_A = 0
    total_B = 0
    total_C = 0

    for row in archivo:
        if row[0] == 'Producto A':
            total_A += int(row[2])
        elif row[0] == 'Producto B':
            total_B += int(row[2])
        else:
            total_C += int(row[2])

with open('data/TotalVentas.json', mode='w', encoding='utf-8') as archivo_json:
    totales = {'Total A': total_A, 'Total B': total_B, 'Total C': total_C}
    json.dump(totales, archivo_json, indent=4, ensure_ascii=False)

root = ET.Element('TotalVentas')

producto_a = ET.SubElement(root, 'ProductoA')
producto_a.text = str(total_A)

producto_b = ET.SubElement(root, 'ProductoB')
producto_b.text = str(total_B)

producto_c = ET.SubElement(root, 'ProductoC')
producto_c.text = str(total_C)

tree = ET.ElementTree(root)
tree.write('data/TotalVentas.xml', encoding='utf-8', xml_declaration=True)
```
